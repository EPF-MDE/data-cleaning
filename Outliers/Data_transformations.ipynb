```python
fig, axs = plt.subplots(1,2, figsize=(16,4))
data = np.random.beta(1, 3, 5000)
sns.histplot(data, kde=True, ax=axs[0])
axs[0].set_title('Beta distribution')
tdata = boxcox(data)[0]
axs[1].set_title('Beta distribution\nafter Box-Cox transformation')
sns.histplot(tdata, kde=True, ax=axs[1])
```
