## Missing volume simulation

```python
data
```

```python
simulation = {}
rates = ( 0.1, 0.2, 0.3)
modes = ['MCAR', 'MAR', 'MNAR']
for r in rates:
    rng = np.random.RandomState(42)
    n_miss = int(r*data.shape[0])
    ind_vec = np.zeros_like(data.Volume)
    ind_vec[:n_miss] = 1
    rng.shuffle(ind_vec)
    data.loc[:, 'MCAR_ind'] = ind_vec
    # print(data['MCAR_ind'])
    diam_factor = data.Diameter/(data.Diameter.max())
    # print(diam_factor)
    ind = np.array(list(map(lambda p: rng.binomial(1, p=r*p, size=1), 
    diam_factor.values)))
    data.loc[:,'MAR_ind'] = ind

    data.loc[:,'MNAR_ind'] = 0
    volume_factor = (data.Volume)/(data.Volume.max())
    # print(volume_factor)
    ind = np.array(list(map(lambda p: rng.binomial(1, p=p*r, size=1), volume_factor.values)))
    data.loc[:, 'MNAR_ind'] = ind
    mdata = {}
    for mode in modes:
        # mdata_i = data.filter(regex=r'[A-Z][a-z]').copy()
        mdata_i = data.copy(deep=True)
        # print((data[f'{mode}_ind']==1).sum())
        mdata_i.loc[mdata_i[f'{mode}_ind']==1, 'Volume'] = None
        mdata[mode] = mdata_i
    simulation[r]=mdata
```

```python
fig, axs = plt.subplots(3,3, figsize=(12,12))
for i, r in enumerate(rates):
    for j, mode in enumerate(modes):
        plt.sca(axs[i][j])
        
        data = simulation[r][mode]
        mrate = data.Volume.isna().mean()
        plt.title(f'{mode}, r={r}, rate={mrate*100:0.0f}%')
        sns.histplot(data,y=f'Volume', kde=True)
        
```
